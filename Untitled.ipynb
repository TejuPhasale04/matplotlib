```python
#importing libraris

import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np

```

```python
%matplotlib inline
```

```python
#Generate sin and cos graph
x=np.linspace(0,3*np.pi,50)

Sin_y=np.sin(x)
Cos_y=np.cos(x)

plt.style.use('classic')

plt.figure(figsize=(10,5))

fig=plt.figure()

#Saving plot in local machine
fig.savefig('SinNCos.png')
plt.plot(x,Sin_y,label='Sine')
plt.plot(x,Cos_y,label='Cosin')

plt.title('SIN N COS WAVES')

plt.legend()
plt.tight_layout()
plt.show()
```
